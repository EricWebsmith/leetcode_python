```python
from typing import Optional, List
import graphviz
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def make_graph(node, graph=None):
    if graph is None:
        graph = graphviz.Digraph()
    
    if node is None:
        return 
    
    graph.node(str(node.val), str(node.val))
    
    if node.left:
        make_graph(node.left, graph)
        graph.edge(str(node.val), str(node.left.val))
        
    if node.right:
        make_graph(node.right, graph)
        graph.edge(str(node.val), str(node.right.val))
    
    return graph
```

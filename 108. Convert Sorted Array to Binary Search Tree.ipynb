# 108. Convert Sorted Array to Binary Search Tree

https://leetcode.com/problems/convert-sorted-array-to-binary-search-tree/

```python
from typing import List, Optional
import graphviz

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def make_graph(node, graph=None):
    if graph is None:
        graph = graphviz.Digraph()
    
    if node is None:
        return 
    
    graph.node(str(node.val), str(node.val))
    
    if node.left:
        make_graph(node.left, graph)
        graph.edge(str(node.val), str(node.left.val))
        
    if node.right:
        make_graph(node.right, graph)
        graph.edge(str(node.val), str(node.right.val))
    
    return graph
```

```python
def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
    middle = len(nums) // 2
    # create root node
    root = TreeNode(nums[middle])
    # Create left tree
    current_index = middle - 1
    current_root = root
    while current_index >= 0:
        current_node = TreeNode(nums[current_index])
        current_root.left = current_node
        current_root = current_node
        current_index -= 1
    
    # Create right tree
    current_index = middle + 1
    current_root = root
    while current_index < len(nums):
        current_node = TreeNode(nums[current_index])
        current_root.right = current_node
        current_root = current_node
        current_index += 1
    
    return root
```

```python
class Solution:
    def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
        n = len(nums)
        middle = n //2
        if n == 0:
            return None
        if n == 1:
            return TreeNode(nums[0])
        
        root = TreeNode(nums[middle])
        
        print(f"----------middle {nums[middle]}----------")
        # Create left tree
        if n > 2:
            root.left = self.sortedArrayToBST(nums[0: middle])
            print(f"left {nums[middle:]}")
        if n > 1:
            root.right = self.sortedArrayToBST(nums[middle+1:])
            print(f"right {nums[middle:]}")
        return root
```

```python
def sorted_array2bst(nums: List[int]) -> Optional[TreeNode]:
    n = len(nums)
    middle = n //2
    if n == 0:
        return None
    if n == 1:
        return TreeNode(nums[0])
    root = TreeNode(nums[middle])
    root.left = sorted_array2bst(nums[0: middle])
    root.right = sorted_array2bst(nums[middle+1:])
    return root
```

```python
nums = [-10,-3,0,5,9]
root = sorted_array2bst(nums)
```

```python
graph = make_graph(root)
display(graph)
```
